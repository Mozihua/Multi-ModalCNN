```python
from keras.preprocessing.image import ImageDataGenerator
from keras import models, layers, optimizers, callbacks
from keras.layers.convolutional import Conv2D, MaxPooling2D, ZeroPadding2D
from keras.layers.core import Flatten, Dense, Dropout
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from keras.utils import to_categorical
from keras import backend as K
```

```python
train_idg = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        vertical_flip=True
    )
val_idg = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        vertical_flip=True
    )
test_idg = ImageDataGenerator()


train_g = train_idg.flow_from_directory(
    '/media/avell/Novo volume/GM_AXIAL/fold_0/train',
    target_size=(91,109),
    batch_size= 10,
    class_mode='binary')
val_g   = val_idg.flow_from_directory(
    '/media/avell/Novo volume/GM_AXIAL/fold_0/validation',
    target_size=(91,109),
    batch_size= 10,
    class_mode='binary')
test_g = test_idg.flow_from_directory(
    '/media/avell/Novo volume/GM_AXIAL/fold_0/test',
    target_size=(91,109),
    batch_size= 10,
    class_mode='binary')
```

```python
K.clear_session()
# CNN Architecturem for VGG16
model = models.Sequential()
model.add(ZeroPadding2D((1,1),input_shape=(91, 109, 3)))
model.add(Conv2D(64, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(64, 3, 3, activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(128, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(128, 3, 3, activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(256, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(256, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(256, 3, 3, activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, 3, 3, activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

# model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, 3, 3, activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(Flatten())
model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

print(model.summary())

# Model Loss function and Optimizer method
compile = model.compile(optimizer=optimizers.SGD(lr=1e-5), loss='binary_crossentropy',
                           metrics=['accuracy'])

# Settting Callbacks
check_p = callbacks.ModelCheckpoint(filepath='axialz_cnn_{acc:.2f}.h5',
                                    monitor='acc', verbose=1,
                                    save_best_only=True, save_weights_only=False)

reduce_lr = callbacks.ReduceLROnPlateau(monitor='acc', factor=0.95, patience=3,
                                       verbose=1, cooldown=2)
callb_l = [check_p, reduce_lr]


# Training Options
fit = model.fit_generator(generator=train_g, steps_per_epoch=train_g.samples/train_g.batch_size, 
                          epochs=100, verbose=1, callbacks=callb_l,validation_data=val_g)

# Saving Model
model.save(filepath=r'axialz_cnn.h5', overwrite=True)
```

```python
#CNN Archtitecture for AlexNet
    
model = models.Sequential()
model.add(ZeroPadding2D((1,1),input_shape=(y, z, 1)))
model.add(Conv2D(64, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(64, 3, 3, activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(128, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(128, 3, 3, activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(256, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(256, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(256, 3, 3, activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, 3, 3, activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, 3, 3, activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, 3, 3, activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(Flatten())
model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2, activation='sigmoid'))

print(model.summary())

# Model Loss function and Optimizer method
sgd = optimizers.SGD(lr=0.0001, decay=1e-6, momentum=0.9, nesterov=True)
compile = model.compile(optimizer=sgd, loss='binary_crossentropy',
                           metrics=['accuracy'])

# Settting Callbacks
check_p = callbacks.ModelCheckpoint(filepath='coronal_cnn_{acc:.2f}.h5',
                                    monitor='acc', verbose=1,
                                    save_best_only=True, save_weights_only=False)
reduce_lr = callbacks.ReduceLROnPlateau(monitor='acc', factor=0.95, patience=3,
                                        verbose=1, cooldown=2)
callb_l = [check_p, reduce_lr]

# Training Options
fit = model.fit_generator(generator=train_g, steps_per_epoch=50, epochs=100, verbose=1,
                             callbacks=callb_l, validation_data=val_g, validation_steps=4)

# Saving Model
model.save(filepath=r'coronal_cnn.h5', overwrite=True)
```
